```python
from colour import Color

def sc_palette(n):
    color_space = [
        "#E41A1C", "#377EB8", "#4DAF4A", "#984EA3", "#F29403", 
        "#F781BF", "#BC9DCC", "#A65628", "#54B0E4", "#222F75", 
        "#1B9E77", "#B2DF8A", "#E3BE00", "#FB9A99", "#E7298A", 
        "#910241", "#00CDD1", "#A6CEE3", "#CE1261", "#5E4FA2", 
        "#8CA77B", "#00441B", "#DEDC00", "#B3DE69", "#8DD3C7", 
        "#999999"
    ]
    
    if n <= len(color_space):
        colors = color_space[:n]
    else:
        start_colors = [Color(hex=c) for c in color_space]
        colors = [c.hex for c in Color.range_to(start_colors[0], start_colors[-1], n)]    
    return colors

colors = sc_palette(200)
categories = adata_labeled.obs['harmonized_celltype'].unique()
color_dict = dict(zip(categories, colors))
```

```python
# Create a scoring system based on column position
def calculate_gene_scores():
    scores = {}
    columns = gene_sets.columns
    
    # Iterate through each column position
    for col_idx, col_name in enumerate(columns):
        # Get genes in this position across all gene sets
        genes = gene_sets[col_name].dropna()
        
        # Calculate weight (higher weight for earlier positions)
        weight = 1 / (col_idx + 1)  # Simple decay function
        
        # Add weighted scores to each gene
        for gene in genes:
            if gene not in scores:
                scores[gene] = 0
            scores[gene] += weight
    
    # Convert to DataFrame for easier handling
    score_df = pd.DataFrame.from_dict(scores, orient='index', columns=['score'])
    score_df = score_df.sort_values('score', ascending=False)
    
    return score_df

# Calculate scores and get top 5000 genes
gene_scores = calculate_gene_scores()
top_5000_genes = gene_scores.head(5000).index.tolist()

# Print some statistics
print(f"Total unique genes scored: {len(gene_scores)}")
print("\nTop 20 genes and their scores:")
print(gene_scores.head(20))

# Show distribution of selected genes across original gene sets
gene_set_coverage = pd.Series({
    index: len([g for g in row.dropna() if g in set(top_5000_genes)])
    for index, row in gene_sets.iterrows()
})

print("\nNumber of selected genes in each gene set:")
print(gene_set_coverage.sort_values(ascending=False))

# Optionally save the genes to a file
# with open('top_5000_genes.txt', 'w') as f:
#     for gene in top_5000_genes:
#         f.write(f"{gene}\n")
```

```python
def calculate_gene_scores():
    scores = {}
    columns = gene_sets.columns
    
    # Calculate row lengths (excluding NaN)
    row_lengths = gene_sets.notna().sum(axis=1)
    max_length = row_lengths.max()
    
    # Print row lengths for verification
    print("Row lengths:")
    print(row_lengths)
    print(f"\nLongest row length: {max_length}")
    
    # Iterate through each row
    for idx, row in gene_sets.iterrows():
        # Get valid genes (non-NaN) and their positions
        valid_genes = row.dropna()
        row_length = len(valid_genes)
        
        # Calculate weights for this row
        for pos, gene in enumerate(valid_genes):
            # Normalize position to longest row length
            normalized_pos = pos * max_length / row_length
            # Calculate weight (linear decay from 1 to 0)
            weight = 1 - (normalized_pos / max_length)
            
            # Add weighted score to gene
            if gene not in scores:
                scores[gene] = 0
            scores[gene] += weight
    
    # Convert to DataFrame for easier handling
    score_df = pd.DataFrame.from_dict(scores, orient='index', columns=['score'])
    score_df = score_df.sort_values('score', ascending=False)
    
    return score_df

# Calculate scores and get top 5000 genes
gene_scores = calculate_gene_scores()
top_5000_genes = gene_scores.head(5000).index.tolist()

# Print statistics
print("\nTop 20 genes and their scores:")
print(gene_scores.head(20))

# Show distribution of selected genes across original gene sets
gene_set_coverage = pd.Series({
    index: len([g for g in row.dropna() if g in set(top_5000_genes)])
    for index, row in gene_sets.iterrows()
})

print("\nNumber of selected genes in each gene set:")
print(gene_set_coverage.sort_values(ascending=False))
```

```python
from scipy.stats import median_abs_deviation
def is_outlier(adata, metric: str, nmads: int):
    M = adata.obs[metric]
    outlier = (M < np.median(M) - nmads * median_abs_deviation(M)) | (np.median(M) + nmads * median_abs_deviation(M) < M)
    return outlier

adata.obs["outlier"] = (
    is_outlier(adata, "log1p_total_counts", 5)
    | is_outlier(adata, "log1p_n_genes_by_counts", 5)
    | is_outlier(adata, "pct_counts_in_top_20_genes", 5)
)
adata.obs.outlier.value_counts()
```

```python
adata.obs["mt_outlier"] = is_outlier(adata, "pct_counts_mt", 3) | (adata.obs["pct_counts_mt"] > 8)
adata.obs.mt_outlier.value_counts()
```

```python
def generate_colors_for_categories(adata, obs_key='celltype', max_colors=30):
    # Get unique categories, excluding NaN values
    categories = adata.obs[obs_key].dropna().unique()
    n_categories = len(categories)

    if n_categories > max_colors:
        print(f"Warning: {n_categories} categories found. Limiting to top {max_colors} most frequent.")
        # Get top max_colors most frequent categories
        top_categories = (adata.obs[obs_key].value_counts().nlargest(max_colors).index.tolist())
        categories = top_categories
        n_categories = len(categories)

    # Generate colors using a colormap
    cmap = plt.get_cmap('tab20')  # This gives 20 distinct colors
    colors = [cmap(i) for i in np.linspace(0, 1, n_categories)]
    
    # Create a dictionary mapping categories to colors
    color_dict = dict(zip(categories, [plt.matplotlib.colors.to_hex(c) for c in colors]))
    
    # Add a color for NaN/Other
    color_dict['Other'] = '#808080'  # Grey color for NaN and other categories
    
    # Store the color dictionary in adata.uns
    adata.uns[f'{obs_key}_colors'] = color_dict
    
    # Update adata.obs to replace rare categories and NaNs with 'Other'
    adata.obs[f'{obs_key}_color_group'] = adata.obs[obs_key].apply(lambda x: x if x in color_dict else 'Other')
    
    return adata
```

```python
from matplotlib.colors import LinearSegmentedColormap, to_hex
```

```python
def create_custom_palette(adata, obs_key='celltype', n_colors=20):
    # Get unique categories and their frequencies
    value_counts = adata.obs[obs_key].value_counts()
    categories = value_counts.index
    n_categories = len(categories)

    # Create a custom colormap
    colors = plt.cm.rainbow(np.linspace(0, 1, n_colors))
    custom_cmap = LinearSegmentedColormap.from_list("custom", colors, N=n_colors)

    # Create color dictionary
    color_dict = {}
    for i, category in enumerate(categories):
        color_dict[category] = to_hex(custom_cmap(i % n_colors))

    return color_dict

# Create the custom palette
custom_palette = create_custom_palette(adata, obs_key='celltype', n_colors=30)

# Plot using the custom palette
sc.pl.umap(adata, color="celltype", palette=custom_palette)
```
